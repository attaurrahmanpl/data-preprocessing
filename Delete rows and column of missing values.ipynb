```python
clean_data = data.drop([ 'LotFrontage' , 'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'] , axis=1)
```

```python
# now compare original data and clean data on heat map
plt.figure(figsize = (10 , 20))
plt.subplot(1,2,1)
sns.heatmap(clean_data.isnull())
plt.title('Now we can see there is low missing value after deleting more 17% missing value column')
plt.subplot(1,2,2)
sns.heatmap(data.isnull())
plt.show()
```

```python
# still we have missing values in the data set 
# now we will use second method
# drop all the null value and store in new data frame
clean_data = clean_data.dropna()
```

```python
plt.figure(figsize = (18 , 8))
sns.heatmap(clean_data.isnull())
plt.show()
# now there i no missing vlalues in data set
```

```python
# Define a function to creat data frame of actual and clean data and show that %age of that
def per_AandC(var):
    return pd.concat([data[var].value_counts()/data.shape[0]*100,
          clean_data[var].value_counts()/clean_data.shape[0]*100],axis=1,
         keys=[var + '_real',var + '_clean'])
    
```
